```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("../data/telecom_customer_churn.csv")
```

## Monthly Charge Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(x="Monthly Charge", kde=True, data = df)
plt.title("Monthly Charge Histogram")
plt.savefig("../screenshots/monthly_charges_histogram.png")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(x="Monthly Charge", data=df)
plt.title("Monthly charges Boxplot")
plt.savefig("../screenshots/monthly_charges_boxplot")
plt.show()
```

### Business Insight

The monthly charges distribution appears moderately skewed.

Some customers have significantly higher monthly charges, indicating premium service users or bundled plans.

Outliers may represent high-value customers for the business.
