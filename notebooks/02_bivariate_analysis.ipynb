```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("../data/telecom_customer_churn.csv")
```

## Customer Status vs Tenure in Months

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Customer Status", y="Tenure in Months", data=df)
plt.title("Customer Status vs Tenure in Months")
plt.savefig("../screenshots/customer_status_vs_tenure.png")
plt.show()
```

### Business Insight
Customers with lower tenure are more likely to churn, while long-term customers tend to stay.

## Customer Status vs Contract Type

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Contract", hue="Customer Status", data=df)
plt.title("Customer Status vs Contract Type")
plt.savefig("../screenshots/customer_status_vs_contract.png")
plt.show()
```

### Business Insight

Customers with month-to-month contracts tend to churn more frequently than customers with long-term contracts.

## Customer Status vs Tenure

```python
plt.figure(figsize=(8,5))
sns.histplot(x="Tenure in Months", hue="Customer Status", kde= True,  data = df)
plt.title("Customer Status vs Tenure")
plt.savefig("../screenshots/customer_status_vs_tenure_hist.png")
plt.show()
```

### Business Insight

Customers with shorter tenure are more likely to churn, while long-term customers are more likely to stay.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Customer Status", y="Tenure in Months", data=df)
plt.title("Tenure Distribution by Customer Status")
plt.savefig("../screenshots/customer_status_vs_tenure_boxplot.png")
plt.show()
```

### Business Insight

Customers who remained with the company generally have higher tenure compared to churned customers.

## Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(14,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.savefig("../screenshots/heatmap.png")
plt.show()
```

### Business Insight

Tenure and Total Charges are positively correlated, indicating that long-term customers generate higher revenue.

Monthly Charge also shows relationships with several service-related features.
